```
import pandas as pd

jobs = pd.read_csv("../data/raw/jobs_raw.csv")
skills = pd.read_csv("../data/raw/skills_raw.csv")
```

```
jobs.head()

jobs.info()

jobs.describe(include="all")

jobs.isnull().sum()

jobs.duplicated().sum()
```

```
skills.head()

skills.info()

skills.describe(include="all")

skills.isnull().sum()

skills.duplicated().sum()
```

```
print("Jobs Shape:", jobs.shape)
print("Skills Shape:", skills.shape)

print("\nMissing Values (Jobs):")
print(jobs.isnull().sum())

print("\nMissing Values (Skills):")
print(skills.isnull().sum())

print("\nDuplicate Jobs:", jobs.duplicated().sum())
print("Duplicate Skills:", skills.duplicated().sum())
```

```
jobs.to_csv("../data/processed/jobs_clean.csv", index=False)
skills.to_csv("../data/processed/skills_clean.csv", index=False)
```
