```
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```
jobs = pd.read_csv("../data/processed/jobs_clean.csv")
skills = pd.read_csv("../data/processed/skills_clean.csv")
```

```
print("Jobs Dataset Shape:", jobs.shape)
print("Skills Dataset Shape:", skills.shape)

print("\nJobs Columns:")
print(jobs.columns.tolist())

print("\nSkills Columns:")
print(skills.columns.tolist())
```

```
top_jobs = jobs["job_title"].value_counts().head(10)

print(top_jobs)

plt.figure(figsize=(10,6))
top_jobs.sort_values().plot(kind="barh")
plt.title("Top 10 AI Job Titles")
plt.xlabel("Number of Jobs")
plt.ylabel("Job Title")
plt.show()
```

```
top_countries = jobs["country"].value_counts().head(10)

print(top_countries)

plt.figure(figsize=(10,6))
top_countries.sort_values().plot(kind="barh")
plt.title("Top 10 Hiring Countries")
plt.xlabel("Number of Jobs")
plt.ylabel("Country")
plt.show()
```

```
experience = jobs["experience_level"].value_counts()

print(experience)

plt.figure(figsize=(8,5))
experience.plot(kind="bar")
plt.title("Experience Level Distribution")
plt.xlabel("Experience Level")
plt.ylabel("Count")
plt.show()
```

```
employment = jobs["employment_type"].value_counts()

print(employment)

plt.figure(figsize=(8,5))
employment.plot(kind="bar")
plt.title("Employment Type Distribution")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.show()
```

```
remote = jobs["remote_type"].value_counts()

print(remote)

plt.figure(figsize=(8,5))
remote.plot(kind="bar")
plt.title("Work Mode Distribution")
plt.xlabel("Work Mode")
plt.ylabel("Count")
plt.show()
```

```
company = jobs["company_size"].value_counts()

print(company)

plt.figure(figsize=(8,5))
company.plot(kind="bar")
plt.title("Company Size Distribution")
plt.xlabel("Company Size")
plt.ylabel("Count")
plt.show()
```

```
jobs["average_salary"] = (
    jobs["salary_min_usd"] + jobs["salary_max_usd"]
) / 2

salary_by_role = (
    jobs.groupby("job_title")["average_salary"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print(salary_by_role)

plt.figure(figsize=(10,6))
salary_by_role.sort_values().plot(kind="barh")
plt.title("Top 10 Highest Paying AI Roles")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Job Title")
plt.show()
```

```
top_skills = skills["skill"].value_counts().head(20)

print(top_skills)

plt.figure(figsize=(10,8))
top_skills.sort_values().plot(kind="barh")
plt.title("Top 20 Most Demanded AI Skills")
plt.xlabel("Frequency")
plt.ylabel("Skill")
plt.show()
```

```
skill_categories = skills["skill_category"].value_counts()

print(skill_categories)

plt.figure(figsize=(10,6))
skill_categories.plot(kind="bar")
plt.title("Skill Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```
