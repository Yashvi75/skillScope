```
import pandas as pd
```

```
jobs = pd.read_csv("../data/processed/jobs_clean.csv")
skills = pd.read_csv("../data/processed/skills_clean.csv")

df = jobs.merge(skills, on="job_id")

df["average_salary"] = (
    df["salary_min_usd"] +
    df["salary_max_usd"]
) / 2
```

```
dashboard_jobs = jobs.copy()

dashboard_jobs.to_csv(
    "../dashboards/dashboard_jobs.csv",
    index=False
)
```

```
dashboard_skills = (
    skills["skill"]
    .value_counts()
    .reset_index()
)

dashboard_skills.columns = [
    "Skill",
    "Demand"
]

dashboard_skills.to_csv(
    "../dashboards/dashboard_skills.csv",
    index=False
)
```

```
dashboard_salary = (
    jobs.assign(
        AverageSalary=(
            jobs["salary_min_usd"] +
            jobs["salary_max_usd"]
        ) / 2
    )
    .groupby("job_title")["AverageSalary"]
    .mean()
    .reset_index()
)

dashboard_salary.to_csv(
    "../dashboards/dashboard_salary.csv",
    index=False
)
```

```
dashboard_country = (
    jobs["country"]
    .value_counts()
    .reset_index()
)

dashboard_country.columns = [
    "Country",
    "Jobs"
]

dashboard_country.to_csv(
    "../dashboards/dashboard_country.csv",
    index=False
)
```

```
summary = pd.DataFrame({
    "Metric": [
        "Total Jobs",
        "Total Skills",
        "Countries",
        "Cities",
        "Industries"
    ],
    "Value": [
        len(jobs),
        len(skills),
        jobs["country"].nunique(),
        jobs["city"].nunique(),
        jobs["industry"].nunique()
    ]
})

summary.to_csv(
    "../dashboards/dashboard_summary.csv",
    index=False
)
```
